```python
import happybase

print(f"happybase version : {happybase.__version__}")
```

```python
# HBase Thrift Server의 호스트명, 포트번호
hb_conn = happybase.Connection('daou-bd-r01a01', 9090)
hb_table = hb_conn.table("nutch")
hb_table
```

```python
data = [value for key, value in hb_table.scan(filter="SingleColumnValueFilter('CF01','PUBDATE',=,'regexstring:^20200723+',true, true)")]
len(data)
```
